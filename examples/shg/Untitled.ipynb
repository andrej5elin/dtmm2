```python
import matplotlib.pyplot as plt
import numpy as np
import glob

fnames = glob.glob('pol_dep_AH/*.png')
step = 2


im = plt.imread(fnames[0])
shape = im.shape
n = len(fnames)
```

```python
images = np.zeros(shape = (n,)+shape)

for i,fname in enumerate(fnames):
    im = plt.imread(fname)
    images[i] = im
```

```python
argmax = np.argmax(images, axis = 0)
anglemax = argmax*step

plt.imshow(anglemax[400:800,750:1250], cmap = 'twilight')
plt.colorbar()
```

```python
imax = np.zeros(shape)
for i in range(shape[0]):
    for j in range(shape[1]):
        imax[i,j] = images[argmax[i,j],i,j]
plt.imshow(imax, vmin = 0, vmax =0.5)
plt.colorbar()
```
